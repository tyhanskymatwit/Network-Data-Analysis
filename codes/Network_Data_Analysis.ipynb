```python
import matplotlib.pyplot as plt

segments = ['Entertainment', 'Cable networks', 'Publishing', 'Local Media', 'Corporate/Eliminations']
revenue = [10178, 2204, 825, 1830, -523]  # in million USD

plt.figure(figsize=(10, 6))
plt.bar(segments, revenue, color=['blue', 'green', 'orange', 'red', 'purple'])
plt.title('CBS Corporation Revenue by Segment (2018)')
plt.xlabel('Segment')
plt.ylabel('Revenue (Million USD)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
years = [2021, 2022, 2023, 2024]
streaming_revenue = [51.5, 62.3, 76, 95.4]  # in billion USD
traditional_revenue = [157.5, 154.3, 139.5, 132]  # in billion USD

plt.figure(figsize=(10, 6))
plt.plot(years, streaming_revenue, marker='o', label='Streaming TV', color='blue')
plt.plot(years, traditional_revenue, marker='o', label='Traditional Pay TV', color='red')
plt.title('Revenue of Traditional Pay TV vs. Video Streaming (2021-2024)')
plt.xlabel('Year')
plt.ylabel('Revenue (Billion USD)')
plt.legend()
plt.grid(linestyle='--', alpha=0.7)
plt.show()
```

```python
# United States (White respondents)
us_genres = ['Comedy', 'Action', 'Thriller', 'Drama', 'Science/Nature']
us_popularity = [90, 88, 83, 81, 80]  # in %

# Australia (Female respondents)
au_genres = ['News', 'Reality TV', 'Dramas', 'Comedies', 'Documentaries']
au_popularity = [62, 45, 37, 30, 31]  # in %

# Japan
jp_genres = ['Variety', 'News', 'Domestic Drama', 'Anime', 'Documentary']
jp_popularity = [37, 27.5, 24.1, 20.1, 19.4]  # in %

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.barh(us_genres, us_popularity, color='blue')
plt.title('United States (2018)')
plt.xlabel('Popularity (%)')

plt.subplot(1, 3, 2)
plt.barh(au_genres, au_popularity, color='green')
plt.title('Australia (2018)')
plt.xlabel('Popularity (%)')

plt.subplot(1, 3, 3)
plt.barh(jp_genres, jp_popularity, color='orange')
plt.title('Japan (2021)')
plt.xlabel('Popularity (%)')

plt.suptitle('Popular TV Genres by Region')
plt.tight_layout()
plt.show()
```
